```python
print(df['BMI'])  
print(df[['BMI', 'SkinCancer']]) 
```

```python
filtered_df = df[df['SkinCancer'] == "Yes"] 
print(filtered_df)
```

```python
df['BMI_Category'] = ['Underweight' if x < 18.5 else 'Normal' if x < 25 else 'Overweight' for x in df['BMI']]
print(df[['BMI', 'BMI_Category']].head())
```

```python
df_dropped = df.drop(['BMI_Category'], axis=1)
df_dropped
```

```python
sorted_df = df.sort_values(by='BMI', ascending=False)
sorted_df
```

```python
df['BMI'] = df['BMI'].fillna(df['BMI'].mean()) 
df.dropna(subset=['PhysicalHealth'], inplace=True)
df
```

```python
grouped = df.groupby('Sex')['BMI'].mean()
print(grouped)
```

```python
df1 = df[['Sex', 'BMI']]
df2 = df[['Sex', 'SkinCancer']]
merged_df = pd.merge(df1, df2, on='Sex')
merged_df
```

```python
plt.plot(df['BMI'][:50], color='blue', label='BMI Trend')
plt.xlabel('BMI SAMPLES')
plt.ylabel('BMI')
plt.title('BMI Line Plot')
plt.legend()
plt.show()
```

```python
df['SkinCancer'].value_counts().plot(kind='bar', color='green')
plt.title('Skin Cancer Occurrences')
plt.show()
```

```python
plt.hist(df['BMI'], bins=10, color='purple', alpha=0.7)
plt.title('BMI Distribution')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.scatter(df['PhysicalHealth'], df['MentalHealth'], color='red')
plt.title('Physical vs Mental Health')
plt.xlabel('PhysicalHealth')
plt.ylabel('MentalHealth')
plt.show()
```

```python
df['Sex'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, shadow=True)
plt.title('Gender Distribution')
plt.show()
```
